### Import copy

Allows to deep copy or shallow copy mutable object using the inbuilt functions present in the module “copy“

```python
import copy
```

### Fn to take Input
Takes as input the size of the chess board - minimum 4x4

```python
def take_input():

    while True:
        try:
            size = int(input('What is the size of the chessboard? n = \n'))
            if size == 1:
                print("Trivial solution, choose a board size of atleast 4")
            if size <= 3:
                print("Enter a value such that size>=4")
                continue
            return size
        except ValueError:
            print("Invalid value entered. Enter again")
```

### Fn to make the board

```python
def get_board(size):
    board = [0]*size
    for ix in range(size):
        board[ix] = [0]*size
    return board
```

### Fn to check for feasability of board/backtrack if not

```python
def solve(board, col, size):
    #base case
    if col >= size:
        return
    
    for i in range(size):
        if is_safe(board, i, col, size):
            board[i][col] = 1
            if col == size-1:
                add_solution(board)
                board[i][col] = 0
                return
            solve(board, col+1, size)
            #backtrack
            board[i][col] = 0
```

### Fn to check for feasable solutions

```python
def is_safe(board, row, col, size):
    #check row on left side
    for iy in range(col):
        if board[row][iy] == 1:
            return False
    
    ix, iy = row, col
    while ix >= 0 and iy >= 0:
        if board[ix][iy] == 1:
            return False
        ix-=1
        iy-=1
    
    jx, jy = row,col
    while jx < size and jy >= 0:
        if board[jx][jy] == 1:
            return False
        jx+=1
        jy-=1
    
    return True
```

### Fn to add feasible board to list of solutions

```python
def add_solution(board):
    global solutions
    saved_board = copy.deepcopy(board)
    solutions.append(saved_board)
```

### Fn to print all the solutions

```python
def print_solutions(solutions, size):
    for sol in solutions:
        for row in sol:
            print(row)
        print()
```

### Main control

```python
size = take_input()

board = get_board(size)

solutions = []

solve(board, 0, size)

print_solutions(solutions, size)

print("Total solutions = {}".format(len(solutions)))
```
